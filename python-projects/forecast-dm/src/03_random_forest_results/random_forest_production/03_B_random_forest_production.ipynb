```python
#####################################################
#
# APLICAR Random Forest a datos NUEVOS preprocesados con PCA
#
#####################################################
# Deben cargarse los archivos
# - T_new_final.csv (son los datos nuevos ya preprocesados con PCA)
# - expected_columns.json
# - modelo_random_forest.pkl
#
# Devolverá
# Random_forest_nuevos_predicciones.csv: csv de T_new_final con las predicciones del random forest
#####################################################

# ===== Carga del modelo y predicción en datos nuevos =====
import pandas as pd
import joblib
import json

# Cargar artefactos
########################################################################
modelo = joblib.load("../random_forest/modelo_random_forest.pkl")
with open("../random_forest/expected_columns.json", "r", encoding="utf-8") as f:
    ########################################################################
    expected_cols = json.load(f)["columns"]

# Usar en un lote nuevo
#################################################################################################
df_nuevo = pd.read_csv("../../01_preprocessing_results/preprocessing_production/T_new_final.csv")
##################################################################################################
y_pred = modelo.predict(df_nuevo)

out = df_nuevo.copy()
out["yhat"] = y_pred
out.to_csv("Random_forest_nuevos_predicciones.csv", index=False)
print("Predictions saved: Random_forest_nuevos_predicciones.csv")
```
